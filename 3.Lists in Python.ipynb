```python
# len(), index(),count(), max(), min(), sum() function 

list2 = [1,2,3,4,4,4,5,6,10,0,-1,100.25,True]
print("Original List Value:",list2)

# total value inside the list
print("Len:",len(list2))

print("Index:",list2.index(3))
print("Index:",list2.index(10))
print("Count:",list2.count(4))
print("Count:",list2.count(10))
print("Max:",max(list2))
print("Min:",min(list2))
print("sum of the list:",sum(list2))
```
