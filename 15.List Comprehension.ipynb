## 15. Python List comprehension

### 1. What is List Comprehension?

* `List comprehension` is a short and powerful to create lists in Python using a single line of code.

* Instead of writing long loops to create a list, we can create the same logic in a cleaner and faster format.


✔️Definition

* List comprehension is a compact syntax used to:
  * generate a list
  * transform data
  * filter elements
  * combine loops and conditions

### 2. Why List Comprehensions is Used?

* ✅Less Code
* ✅More readable(when used properly)
* ✅Faster than traditional loops
* ✅Used heavily in real-world Python programming
* ✅Best for data transformation and filtering

### 3. Traditional Approach VS List Comprehension

#### Example: Creating a list of squares

### Traditional Approach

```python
squares = []

for i in range(1,11):
    squares.append(i*i)

print(squares)
```

### 4. General Syntax of List Comprehenstion

### List Comprehensions

```python
squares = [i*i for i in range(1,11)]
print(squares)
```

### Breakdown

|Part|Meaning|
|---|---|
|expression|What you want to store in the list|
|item| Variables that takes values|
|iterable|list, tuple, range, string, set, dictionary etc|

### 5. List Comprehension with Condition(Filtering)

```python
# Example : display only even numbers from 1-10

# Traditional method 
even = []
for i in range(1,11):
    if i%2==0:
        even.append(i)

print(even)
```

```python
even = [i for i in range(1,11) if i%2==0]
even
```

* Loop runs from 1 to 10
* condition check i%2==0
* only even numbers get stored

### 6. List Comprehension with if-else(Condition Expression)

Important Notes:

When using if-else, the syntax changes

#### syntax

newList = [expression_if_true if condition else expression_if_false for item in iterable]

newList

### Example: Replace odd numbers with 0

```python
result = [i if i%2==0 else 0 for i in range(1,11)]
result
```

```python
result = [i if i%2==0 else "odd" for i in range(1,11)]
result
```

```python
[i for i in range(1,11) if i%2==0]  # wrong method
```

```python
[i if i%2==0 else 0 for i in range(1,11) ]
```

####  points 

* if condition at the end = filtering
* if-else in the beginning = condition assignment 

### 8. Nested List Comprehenstions(Multiple Loops)

newList = [expression for item1 in iterable1 for item2 in iterable2]

newList

```python
### Example: Generate pairs

pairs = [(x,y) for x in [1,2,3] for y in [10,20]]
pairs
```

```python
pairs = []

for x in [1,2,3]:
    for y in [10,20]:
        pairs.append((x,y))
print(pairs)
```

### List Comprehension with Strings

```python
vowels = [ch for ch in str1 if ch in "aeiou"]
print(vowels)
```

### List Comprehension with strip()

### Example: Removes spaces from user input list 

```python
dataList = ['   Ravi   ',"  Ajay","Kiran   ",",   ","  John"]
print(dataList)
```

```python
cleanNames= [name.strip() for name in dataList]
cleanNames
```

### List Comprehension with split()



```python
sentence = "Python is very easy"
words = [word for word in sentence.split()]
words
```

### List Comprehension with Type Conversion

##### Example: Convert list of strings to integers


```python
nums = ["10","20","30","40"]
print(nums)
int_nums = [int(n) for n in nums]
int_nums
```

### List Comprehension with Dictionary

##### Example:Extract key

```python
data = {"name":"John","age":25,"city":"Delhi"}

keys = [key for key in data]
print(keys)
```

```python
values = [value for value in data.values()]
values
```

### List comprehension with enumerate()

##### Example: Get index and value




```python
items = ['apple',"banana","mango"]


result  = [(i, item) for i,item in enumerate(items)]
result
```

```python
data = {"name":"John","age":25,"city":"Delhi"}

result1 = [(key,value) for key,value in data.items()]
result1
```

### List Comprehension with zip()

#### Example: Combine two lists


```python
names = ["John","Alice","Bob"]
scores = [90,100,95]


combined = [(n,s) for n,s in zip(names, scores)]
combined
```

#### Using List Comprehension with Logical Condition

##### Example: Filter numbers greater than 50

```python
marks = [34,56,78,90,45,100]

passed = [m for m in marks if m>50]
passed
```

#### List Comprehension with Multiple Conditions 

##### Example: Filter even numbers and divisible by 3

```python
result = [x for x in range(1,51) if x%2==0 and x%3==0 ]
result
```
