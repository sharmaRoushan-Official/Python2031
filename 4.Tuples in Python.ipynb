# 4. Tuples in Python

 * __Tuples are immutable lists and cannot be changed in any way once it is created.__
 * __Tuples are defined in the same way as lists.__
 * __They are enclosed within parenthesis and not within square braces.__
 * __Tuples are ordered, indexed collections of data.__
 * __Similar to string indices, the first value in the tuple will have the index[0], the second value [1].__
 * __Negative indices are counted from the end of the tuple, just like lists.__
 * __Tuples also have the same strcucture where commas separated the values.__
 * __Tuples can sotre duplicate values.__
 * __Tuples allow you to store several data items including string, integer, float in one variable.__

```python
# checking Tuple type


data = (1,2,3,4,5,6,7)
print(type(data))
```

```python
# negative indexing value
print(data[-1])
print(data[-2])
print(data[-3])
print(data[-4])
print(data[-5])
print(data[-6])
print(data[-7])
```

```python
# Slicing of tuples

nums = (10,20,30,40,50,60,70,"Hello world",True,False)
print(nums,type(nums))
```

```python
# Striding value of tuples
print(nums)
x =nums[::2]
print(x)
```

```python
# Tuple function 

# 1. len()

t = (5,10,15)
print(len(t))
```
