## type()

```python
# String value

print("Hello world")
print(type("Hello world"))
print("Python")
print(type("Python"))
print("100")
print(type("100"))
print('100.23')
print(type('100.23'))


# integer
print(100)
print(type(100))
print(1000)
print(type(1000))
print(10)
print(type(10))


# float value
print(100.25)
print(type(100.25))
print(145.366699)
print(type(145.366699))


# Boolean
print(True)
print(type(True))
print(False)
print(type(False))
```
