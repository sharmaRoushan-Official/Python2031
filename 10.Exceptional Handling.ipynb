### 4. Basic Structure of Exception Handling

âœ… Syntax

```python
try:
    a = 10
    b = 0
    print(a/b)
except:  # default except
    print("Error Occurred")

print("Hello world")
```

```python
try:
    a = 10
    b = 0
    print(a/b)
except ZeroDivisionError:  
    print("ZeroDivisionError")

print("Hello world")
```

```python
try:
    a = 10
    b = 0
    print(a1/b)
except Exception as e:  
    print(e)

print("Hello world")
```

### 5. Handling Specific Exceptions

except:

```python
try:
    a = int(input("Enter first value: "))
    b = int(input("Enter second value: "))
    print(a/b)
except ZeroDivisionError:
    print("Cannot divide by Zero")
except ValueError:
    print("Please enter valid numbers.")
```

## 7. Multiple Except Blocks

```python
try:
    x = int("abc")
except ValueError:
    print("Value Error Occurred")
except TypeError:
    print("Type Error Occurred")
```

## 8. Catching Multple Exceptions Together

```python
try:
    a = 10/0
except (ZeroDivisionError, ValueError):
    print("Error")
```

## 9. The else Block

âœ… When does else execute?

 * __Execute only if No Exceptions Occurs__

```python
try:
    a = 10
    b = 2
    print(a/b)

except ZeroDivisionError:
    print("DivisionError")
else:
    print("Exceution successful")
```

## 10. The finally Block

 * __Excutes always__
 * __Used for cleanup(closing files,DB connection)__

```python
try:
    file = open("C:\\Users\\User\\Desktop\\NursaryRhymes.txt",'r')
    print(file.read())
except FileNotFoundError:
    print("File Not Found")
finally:
    print("Closing File")
```

## 11. try-except-else-finally flow

```python
try:
    print("Try Block")
except:
    print("Except Block")
else:
    print("Else Block")
finally:
    print("Finally Block")
```
