```python
# Format 

txt = "Hello {word} word"
print(txt)

print(txt.format(word="Python"))
print(txt.format(word="Yash"))
```

```python
msg1 = "hi,My name is {} and I am {} years old."

print(msg1.format("Bob",44))
print(msg1.format(44,"Bob"))


msg2 = "Hi, my name is {name} and I am {number} years old."
print(msg2.format(name="Bob",number=44))
print(msg2.format(number=44,name="Bob"))

msg3 = "Hi, myself {0} and I am {1} years old."

print(msg3.format("Bob",44))
print(msg3.format(44,"Bob"))
```
