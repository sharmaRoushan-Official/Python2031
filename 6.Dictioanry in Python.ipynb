# 6. Dictionary in Python

 * __A dictionary stores data in key:value pairs__
 * __Keys must be unique and immutable__
 * __Values can be any data type.__
 * __Dictionary is mutable.__
 * __Written using curly braces {}.__

```python
x1 = {}
print(x1,type(x1))
```

```python
# Creating Dictionary

student = {
    "name":"Rahul",
    "age":21,
    "course":"Python"}
print(student,type(student))
```

```python
# dict() constructor

emp = dict(id=102,name="Neha",salary=40000)
print(emp,type(emp))
```

```python
# Accessing Dictionary Values

# using key

print(emp['id'])
print(emp['name'])
print(emp['salary'])
print(emp['missile'])
```

```python
print(emp['salary'])
```

```python
# using get() method : safest.

print(emp.get("id"))
print(emp.get("name"))
print(emp.get("salary"))
print(emp.get('missile'))
```

### Dictionary Methods(Very Important)

|Method|Description|
|-----|--------|
|keys()| Retunrs all keys|
|values()|Returns all values|
|items()| Returns key-values pairs|
|update()| Merge dictionaries|
|copy()| Copy dictionary|

```python
dict1 = {'a':'apple',"b":'boy'}
dict2 = dict(c='cat',d='dog')

dict1.update(dict2)
```

# Looping Through Dictionary

## 1. Loop through keys

```python
dict2
```

```python
for key in dict2:
    print(key)
```

# 2.Loop through values

```python
for value in dict2.values():
    print(value)
```

```python
for value in dict2.keys():
    print(value)
```

```python
for value in dict2.items():
    print(value)
```
