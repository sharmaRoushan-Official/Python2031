## 16. Decorator in Python

* A `Python decorator` is a function that modifies or extends the behavior or another function(or class) without changing its actual code.

* Decorators are very important because they are used heavily in:

  * Django/Flask(authentication,permission)
  * Logging
  * Timing Functions
  * Validation
  * Role-based access
  * Caching

✅ 1.Basic Idea of Decorator

 * In Python, functions are first-class objects, meaning:
   * You can pass a function as argument
   * You can return a function from another fucntion.


### Example


```python
def outer():
    print("outer function running")
    def inner():
        print("Inner funciton running")
    inner()
outer()
```

### 2. Function as Parameter

```python
def greet():
    print('Hello')
    
def call_function(func):   #(greet)
    func()  # greet()
call_function(greet)
```

### Returning Function Example

```python
def outer():
    def inner():
        print("Inner function Executed")
    return inner
f = outer()
f()
```

### 4. Decorator Concept

* A decorator is basically:
  * A function that takes another function
  * Adds extra code
  * Returns the modified function

### 5. Decorator Without @ Syntax

```python
def decorator(x):  # (hello)
    def wrapper():
        print("Before Function call")
        x()  # hello() ---> Hello world
        print('After Function call')
    return wrapper

def hello():
    print("Hello world")

h = decorator(hello)
h()
```

### Decorator With @ Syntax(Most common)

```python
def decorator(x):  # (hello)
    def wrapper():
        print("Before Function call")
        x()  # hello() ---> Hello world
        print('After Function call')
    return wrapper
```

```python
@decorator
def hello():
    print("Hello world")

hello()
```

```python
@decorator
def name():
    x = input("Enter your name:")
    print("congratulations "+ x)
name()
```

### Decorator with *args and **kwargs


```python
def decorator(func):  # (hello)
    def wrapper(*args,**kwargs):
        print(args,kwargs)
        print("Before Function call")
        
        result = func(*args, **kwargs)
        
        print('After Function call')
    return wrapper
```

```python
@decorator
def add(a,b):
    print(a+b)
add(10,20)
```

```python
def decorator(func):  # (hello)
    def wrapper(*args, **kwargs):
        print("Function is starting")
        value = func(*args, **kwargs)
        print("Funciton is Finished")
        print(value)
    return wrapper
```

```python
@decorator
def multiplication(a,b):
    return a+b
multiplication(10,20)
```

```python
import time
# print(time.time())


def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args,**kwargs)
        end = time.time()
        print(f"Execution Time:{end-start} Second")
        return result
    return wrapper
```

```python
@timer
def big_task():
    time.sleep(3)
    print("Task Done")
big_task()
```

```python
@timer
def add(a,b):
    print(a+b)
add(10,20)
```
