## 16. Decorator in Python

* A `Python decorator` is a function that modifies or extends the behavior or another function(or class) without changing its actual code.

* Decorators are very important because they are used heavily in:

  * Django/Flask(authentication,permission)
  * Logging
  * Timing Functions
  * Validation
  * Role-based access
  * Caching

✅ 1.Basic Idea of Decorator

   * You can pass a function as argument
   * You can return a function from another fucntion.

### Decorator With @ Syntax(Most common)

```python
def decorator(x):  # (hello)
    def wrapper():
        print("Before Function call")
        x()  # hello() ---> Hello world
        print('After Function call')
    return wrapper
```

```python
@decorator
def hello():
    print("Hello world")

hello()
```

```python
@decorator
def name():
    x = input("Enter your name:")
    print("congratulations "+ x)
name()
```

### Decorator with *args and **kwargs


```python
def decorator(func):  # (hello)
    def wrapper(*args,**kwargs):
        print(args,kwargs)
        print("Before Function call")
        
        result = func(*args, **kwargs)
        
        print('After Function call')
    return wrapper
```

```python
@decorator
def add(a,b):
    print(a+b)
add(10,20)
```

```python
def decorator(func):  # (hello)
    def wrapper(*args, **kwargs):
        print("Function is starting")
        value = func(*args, **kwargs)
        print("Funciton is Finished")
        print(value)
    return wrapper
```

```python
@decorator
def multiplication(a,b):
    return a+b
multiplication(10,20)
```

```python
import time
# print(time.time())


def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args,**kwargs)
        end = time.time()
        print(f"Execution Time:{end-start} Second")
        return result
    return wrapper
```

```python
@timer
def big_task():
    time.sleep(3)
    print("Task Done")
big_task()
```

```python
@timer
def add(a,b):
    print(a+b)
add(10,20)
```

```python
### Real use Case:

def login_required(func):  # dashboard
    def wrapper(user):   # admin
        if user == "admin":
            return func(user)  # dashboard("admin")
        else:
            return "Access Denied"
    return wrapper
```

```python
def dashboard(user):
    return f"Welcome to dashboard {user}"

x = dashboard("Shyam")
x
```

```python
## Multiple Decorators

def upper_case(func):
    def wrapper():
        return func().upper()
    return wrapper

def excelamation(func):
    def wrapper():
        return func()+ "!!!"
    return wrapper
```

```python
@excelamation
@upper_case
def message():
    userName = input("Enter your name")
    return "Congratulation " + userName

print(message())
```

### Class Decorator(Rare but useful)

```python
def class_decorator(cls):
    cls.company = "TCS"
    return cls

@class_decorator
class Employee:
    def __init__(self,name):
        self.name = name

e = Employee("Ravi")
print(e.name)
print(e.company)
```

## Method Decorator in Class 

```python
from functools import wraps


def check_login(func):
    @wraps(func)
    def wrapper(self,*args,**kwargs):
        if self.logged_in:
            return func(self,*args,**kwargs)
        return "Login Required"
    return wrapper
```

```python
class User:
    def __init__(self,logged_in):
        self.logged_in = logged_in
        
    @check_login
    def profile(self):
        return "Welcome to Profile"

u1 = User(True)
print(u1.profile())

u2 = User(False)
print(u2.profile())
```
