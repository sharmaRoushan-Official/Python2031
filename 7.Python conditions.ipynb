```python
# write a python program to check wheter a trinagle is equilateral|isosceles|or scalen

side1=int(input("Enter first side of trinagle: "))
side2=int(input("Enter second side of tringle: "))
side3=int(input("Enter third side of traingle: "))


# (\(a+b>c\), \(a+c>b\), \(b+c>a\))

if (side1+side2>side3) and (side1+side3>side2) and (side2+side3>side1):
    print(f"Triangle is valid")
    if (side1==side2==side3):
        print(f"This Triangle is Equilateral:{side1}|{side2}|{side3}")
    elif (side1 == side2) or (side1 == side3) or (side2 == side3):
        print(f"This Triangle is Isosceles:{side1}|{side2}|{side3}")
    else:
        print(f"This Triangle is Scalen:{side1}|{side2}|{side3}")
```

```python
ascii_lowercase = 'abcdefghijklmnopqrstuvwxyz'
ascii_uppercase = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
digits = '0123456789'
hexdigits = '0123456789abcdefABCDEF'
octdigits = '01234567'

punctuation = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
whitespace = ' \t\n\r\x0b\x0c'


char = input("Enter any value: ")

if char in ascii_lowercase:
    print(f"Your value is in lower case: {char}")
    if char in hexdigits:
        print(f"Your value is in Hexa Digits:{char}")
elif char in ascii_uppercase:
    print(f"Your value is in upper case: {char}")
    if char in hexdigits:
        print(f"Your value is in Hexa Digits:{char}")
elif char in digits:
    if int(char)>0 and int(char)<8:
        print(f"Your value is in Octa Digit:{char}")
    if char in hexdigits:
        print(f"Your value is in Hexa Digits:{char}")
    print(f"Your value is in digit:{char}")
elif char in punctuation:
    print(f"Your value is in Punctuation:{char}")
elif char in whitespace:
    print(f"Your value is in whitespace:{char}")
```

```python
# Write a Python program to classify *BMI (Body Mass Index)*:

#     # * <18.5 → Underweight
#     # * 18.5–24.9 → Normal
#     # * 25–29.9 → Overweight
#     # * ≥30 → Obese


weight = float(input("Enter your weight in Kg: "))
height = float(input("Enter your heighin in meter: "))


if weight>0 and height>0:
    print("Are you sure this is correct value:")
    con = input("Enter yes/no or Y/N: ").lower()
    if con == "yes" or con == "y":
        bmi = (weight / (height)**2)
        if bmi>0.0 and bmi<=18.5:
            print(f"Underweight- BMI value is:{bmi}")
        elif bmi>18.5 and bmi<=24.9:
            print(f"Normal-BMI value is:{bmi} ")
        elif bmi>24.9 and bmi<=29.9:
            print(f"Overweight-BMI value is:{bmi}")
        elif bmi>=30:
            print(f"Obese-BMI value is:{bmi}")
```
