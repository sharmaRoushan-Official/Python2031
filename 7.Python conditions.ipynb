```python
marks = 80  # nested if 

if marks>0 and marks<=100:
    if marks>=90:
        print("Grade A")
    elif marks>= 75:
        print("Grade B")
    elif marks>=60:
        print("Grade C")
    else:
        print("Fail")
else:
    print("Enter a Valid Marks.")
```

```python
# nested if statment

age = 16
citizen = False

if age>=18:
    if citizen:
        print("Elegible to True")
    else:
        print("Not Citizen")
else:
    print("Under Age")
```
