```python
# Data Types 

# string
print("Hello world")
print('Hello world')
print("100")
print('123.58')
print("!@##$%%")


# integer
print(100)
print(10000)
print(25000)


# float
print(100.0)
print(0.100)
print(25.50)


# Boolean
print(True)
print(False)
```

```python
# Data Types 

# string
print("Hello world")
print(type("Hello world"))
print('Hello world')
print(type('Hello world'))
print("100")
print(type("100"))

print('123.58')
print(type('123.58'))
print("!@##$%%")
print(type("!@##$%%"))



# integer
print(100)
print(type(100))
print(10000)
print(type(10000))
print(25000)
print(type(25000))


# float
print(100.0)
print(type(100.0))
print(0.100)
print(type(0.100))
print(25.50)
print(type(25.50))


# Boolean
print(True)
print(type(True))
print(False)
print(type(False))
```

```python
# Data typecasting 

val = 100  # string, float
print(val)
print(type(val))

strVal = str(val)
print(strVal,type(strVal))


floatVal = float(val)
print(floatVal,type(floatVal))
```

```python
bool1 = True  # string, integer, float
print("Original Value:",bool1)

print("Original Value type:",type(bool1))


bool2 = False   # string, integer, float
print("Original Value:",bool2)
print("Original Value type:",type(bool2))
```
