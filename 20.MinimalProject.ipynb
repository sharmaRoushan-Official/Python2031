### 20. ATM Project

2. Account Balance
3. Deposite Cash
4. Withdraw cash
5. Change Name
6. Show Account Number
7. Change PIN
8. Exit

```python
import random
class atm:
    def __init__(self,balance=0):
        while True:
            self.name = input("Enter your Name Here: ")
            if self.name.replace(" ","").isalpha():
                break
            else:
                print("Name Should be only in Alphabetical Characters.")
        while True:
            self.accountNumber = input("Enter Your Account Number: ")
            if self.accountNumber.isnumeric() and len(self.accountNumber)==6:
                break
            else:
                print("Enter a Valid Account Number")
        while True:
            self.pin = input("Set Account PIN: ")
            if self.pin.isnumeric() and len(self.pin) == 4:
                break
            else:
                print("Enter 4-Digit Numeric Pin Only.")
        self.balance = balance
        

    # method to show customer details
    def details(self):
        """Display masked account details"""
        str1 = self.accountNumber[:2]
        str2 = self.accountNumber[-2:]
        maskedAccount = str1+"XXXXX"+str2
        print(f"Account Holder Name:{self.name}")
        print(f"Account Number:{maskedAccount}")
        print(f"Current Balance:{self.balance}")

    ## method to check balance
    def checkBalance(self):
        """Check Balance:"""
        pin2 = input("Enter your Current PIN to check balance: ")
        if pin2 == self.pin:
            print(f"Your Current Balance:₹{self.balance}") 
        else:
            print("Invalid PIN\nEnter Correct Current PIN")
        
    ## method to deposite cash
    def depositeCash(self):
        pin2 = input("Enter your Current PIN")
        otpValue = random.randint(1000, 9999)
        print("Your Current OTP is:",otpValue)
        otp = int(input("Enter OTP Value: "))
        if otpValue == otp:
            if pin2 == self.pin:
                depoCash = int(input("Enter Your Amount ₹:"))
                self.balance = self.balance+depoCash
                print("Balance Deposited Successfully!!")
                print(f"Your Updated Balance is ₹:{self.balance}")
            else:
                print("Invalid PIN\nEnter Correct PIN.")
        else:
            print("Invalid OTP\nEnter a Valid OTP")

    ### method to withdraw cash
    def wittCash(self):
        pin2 = input("Enter your Currect PIN: ")
        if pin2 == self.pin:
            wittcash = int(input("Enter your Amount to Withdraw ₹: "))
            if wittcash>=10000:
                otpValue = random.randint(1000, 9999)
                print("Your Current OTP is:",otpValue)
                otp = int(input("Enter your Current OTP: "))
                if otpValue == otp:
                    if self.balance>wittcash:
                        self.balance = self.balance-wittcash
                        print(f'Collect Your Cash ₹:{wittcash}\nYour Updated Balance is:{self.balance}')
                    else:
                        print(f"Insufficient Funds ₹:{self.balance}")
                else:
                    print("Invalid OTP\nEnter a Valid OTP")
            if self.balance>wittcash:
                self.balance = self.balance-wittcash
                print(f'Collect Your Cash ₹:{wittcash}\nYour Updated Balance is:{self.balance}')
            else:
                print(f"Insufficient Funds ₹:{self.balance}")
        else:
            print("Invalid PIN\nEnter Correct PIN.")

    ## method to change name
    def changeName(self):
        pin2 = input("Enter current pin to change Name: ")
        otpValue = random.randint(1000,9999)
        print(f"Your Current OTP is:{otpValue}")
        otp = int(input("Enter Your Generated OTP: "))
        if otpValue == otp:
            if pin2 == self.pin:
                oldName = input("Enter old name here: ")
                if oldName == self.name:
                    newName = input("Enter your new name here: ")
                    if newName.replace(" ","").isalpha():
                        self.name = newName
                    else:
                        print(f"Name should be only in Alphabetical form.")
                else:
                    print("Enter Correct Old Name.")
            else:
                print("Invalid PIN\nEnter Correct PIN")
        else:
            print("Invalid OTP\nEnter a valid OTP")
    # method to show account Number
    def showAccountNumber(self):
        pin2 = input("Enter current pin to show Account Number: ")
        if pin2 == self.pin:
            x = self.accountNumber
            print("Your Account Number is:",x)
        else:
            print("Invalid Current PIN")

    # Method to change PIN
    def changePIN(self):
        pin2 = input("Enter Current PIN: ")
        if pin2 == self.pin:
            print("Your Enter PIN is Matched Successfully.")
            accountNum = input("Enter Your Account Number: ")
            if accountNum == self.accountNumber:
                newPIN = input("Enter New PIN: ")
                if newPIN.isnumeric and len(newPIN) == 4:
                    self.pin = newPIN
                    print(f"Your New PIN is updated Successfully\nYour New Current PIN is: {self.pin}")
                else:
                    print("Set Valid PIN in Numeric with 4-Digit only")
            else:
                print("Enter Valid Current Account Number.")
        else:
            print("Enter Valid Current PIN")
            
    # method -1
    def trans(self):
        print("""
        \n~~~~~~~~~~WELCOME TO MYATM~~~~~~~~~~~~~
        \n1.Account Details:
        \n2.Account Balance:
        \n3.Deposite Cash:
        \n4.Withdraw Cash:
        \n5.Change Name:
        \n6.Show AccountNumber:
        \n7.Change PIN
        \n8.Exit
        \n---------------Choose Options Above-------------------
        """)
        while True:
            try:
                opt=int(input("\nChoose any option:"))
            except:
                print("Error:Choose Correct Option.")
                continue
            else:
                if opt == 1:
                    atm.details(self)
                elif opt == 2:
                    atm.checkBalance(self)
                elif opt == 3:
                    atm.depositeCash(self)
                elif opt == 4:
                    atm.wittCash(self)
                elif opt == 5:
                    atm.changeName(self)
                elif opt == 6:
                    atm.showAccountNumber(self)
                elif opt == 7:
                    atm.changePIN(self)
                elif opt == 8:
                    print("Thank For Using This ATM")
                    break
                    
user = atm()
user.trans()
# print(user.name)
# print(user.accountNumber)
# print(user.pin)
# print(user.balance)
```
