### 17. Python Generator

#### ✅ What is Generator in Python??

* A __generator__ is a special type of function that produces values one at a time, instead of producing all values at once.

* A generator does __not stores all results in memory__ like a list.
* Instead, it generates the next value only when needed.

* This is called:

  
✅ __Lazy Evaluation__

✅ __Memory Efficient Iteration__



✅ __Normal Function VS Generator Function__


⭐ __Normal Function__

A normal function returns a value using return.

Once it returns, the function stops permanently.

```python
def normal():
    return 10
normal()
```

⭐ Generator Function

A generator used yield insted of return.

It pauses exectuion and returmes later if required.

```python
def gen():
    yield 10
print(gen())
```

The generator funciton returns a generator object, not the value.

To get the value:

```python
g = gen()
print(next(g))
```

### ✅ What does yield do?

The yield keyword:

✔️ Produces a value

✔️ Pauses the function states

✔️ Saves local variables

✔️ Resumes from same point when called again.

## ✅ Example: Generator with Multiple Yield

```python
def my_generator():
    yield 1
    yield 2
    yield 3

g = my_generator()
print(next(g))
print(next(g))
print(next(g))
```

### Generator Execution Flow()

```python
def test():
    print('Start')
    yield 100
    print("Middle")
    yield 200
    print("End")
    yield 300

test()
```

```python
g = test()
print(next(g))
```

```python
print(next(g))
```

```python
print(next(g))
```

```python
print(next(g))
```

### 📌 Explanation

* Fist next() runs unitl first yield
* Second next() resumes after first yield
* Third next() resumes after second yield

So generator works like pause-resume machine.

### Generator with Loop 

```python
def numbers(n):
    for i in range(1,n+1):
        yield i

g = numbers(5)
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
# print(next(g))
```

```python
def numbers(n):
    for i in range(1,n+1):
        yield i

g = numbers(5)


for value in g:
    print(value)
```

## ✅ Why Generators are Memory Efficient?

⭐ List Example (Consumes Memory)


```python
def list_numbers(n):
    return [i for i in range(n)]

lst = list_numbers(1000000)  # 1 Million numbers in the memory
print(len(lst))
```

### Generator Example(No memory waste)

```python
def gen_numbers(n):
    for i in range(n):
        yield i

g = gen_numbers(1000000)
print(g)
print(next(g))
print(next(g))
print(next(g))
```

 This doest NOT store all numbers. If produces them one by one.

## List Expression 

```python
lst = [x*x for x in range(5)]
print(lst)
```

## Generator Expression

```python
gen = (x*x for x in range(5))
print(gen)
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

### ✅ Key Difference: List VS Generator Expression

|Feature|List|Generator|
|---|---|---|
|Memory|High|VeryLow|
|Speed|Faster Access|Slightly Slower|
|Storage|Stores all values|Stores only one at a time|
|Use Case| Small dataset| Large dataset|

✅ yield VS retun

|Keyword|Behavior|
|---|---|
|return|Ends function permanently|
|yield|Puses function and resumes later|

## ✅ Generator Functions with 

send()


Generators can accept input dynmacally using send()

```python
def mygen():
    x = yield "hello"
    yield x  # display value
g = mygen()
print(next(g))  # Start generator
print(g.send(50))  # send value
```

### Notes

* First next() start generator
* yield "hello" return "hello"
* send(50) assigs 50 to x

###  ✅ Generator with throw()
You can rasie exception inside generator

```python
def gen():
    try:
        yield 1
        yield 2
    except ValueError:
        yield "ValueError Occurred"

g = gen()
print(next(g))
print(next(g))
print(g.throw(ValueError))
```

## Generator with close()

Stop generator permanently

```python
def gen():
    yield 1
    yield 2
    yield 3

g = gen()
print(next(g))
print(next(g))
g.close()
print(next(g))
```
