### âœ… global keyword

```python
x = 5 # global variable call

def change():
    x = 20 # local variable call
    print(x)
change()
print(x)
```

### ðŸŸ¢ Recursion



```python
# factorial - 5! -  1,2,3,4,5=120 

def factorial():
    n = int(input("Enter any value to check factorial: "))
    fact=1
    for a in range(1,n+1):
        
        fact = fact * a
    print(f"Factorial of Number input Value is {n}:{fact}")
        
        
factorial()
# print(fact)
```
