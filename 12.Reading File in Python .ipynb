### 9. Using with statement(Best Practice)

with automatically close the file.

```python
fileName = "C:\\Users\\User\\Desktop\\NursaryRhymes.txt"

with open(fileName,'r') as f:
    print(f.read())
```

```python
fileName = "C:\\Users\\User\\Desktop\\NursaryRhymes.txt"

with open(fileName,'r') as x:
    print(x.read())
```

```python
fileName = "C:\\Users\\User\\Downloads\\archive (1)\\FastFoodNutritionMenuV2.csv"
print(fileName)
```

```python
with open(fileName,'r') as file:
    reader = csv.reader(file)

    print(reader)  # object path
    count = 0
    for row in reader:
        count+=1
        print("--------",count)
        print(row)
```
