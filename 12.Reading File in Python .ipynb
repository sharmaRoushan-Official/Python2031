## 15. Handling File Exceptions

sometimes file may not exist.

```python
try:
    file = open("C:\\Users\\User\\Desktop\\NursaryRhymes.txt","r") 
    print(file.read())

# except FileNotFoundError:
#     print("Error")
except Exception as e:
    print(e)
finally:
    print("File Compiled Successfully")
```
