 * __Sets are unordered, but you can remove items and add new items [mutable].__

 * __A set itself may be mofified, but the elements contained in the set must be of an immutable type.__
 * __Sets are used to store mutltiple items in single variable.__
 * __You can denote a set with a Pair of curly brackets {}.__

```python
# Set:
    # Union
    # Intersetion
    # Difference
    # symmetric_difference
```

```python
# Union(|) - pipe symbol 

a = {1,2,3} 
b = {3,4,5}

a.union(b)  # keyword
```

```python
print(a | b)  # symbol
```

```python
# Intersection 
a = {1,2,3} 
b = {3,4,5}

a&b
```

```python
# Difference 

# a-b
# b-a

a = {1,2,3}  # 1,2,3
b = {3,4,5}  #3,4,5

# -

print(a-b)  # symbol
print(a.difference(b)) # keyword

print(b-a)  #  3,4,5  -  {4,5}
print(b.difference(a))
```

```python
# symmetric_difference

a = {1,2,3,3}  # 1,2
b = {3,4,5}  # 4,5

print(a.symmetric_difference(b)) # keyword
print(a^b)  # symbol
```

```python
# membership Tesing (in and not in)

s = {10,20,30}

print(20 in s)  # Boolean: True|False
print(20 not in s)

print(-20 in s)
print(-20 not in s)
```

### Set Methods 

|Method|Description|
|----------|----------|
|union()| Combines sets|
|intersection()|Commen elements|
|difference()|Elements not common|
|issubset()| Check subset|
|issuperset()|Check Superset|

```python
a = {1,2}
b = {1,2,3}

print(a.issubset(b)) #
print(a.issuperset(b)) # 

print(b.issuperset(a)) 
```
