## 13. Writing Files in Python

### With Python We can:
 * `Create a file`

 * `Write data to a file`

Example uses:

# 3. File Opening Modes (very Important)
---
|Mode|Meaning|Description|
|---|---|---|
|"r"|Read|Opens file for reading only.|
|"w"|Write|Creates file or overwrites existing file|
|"a"|Append|Adds data at end of the file|
|"x"|Creates|Creates new file, error if exists|
|"b"|Binary|Used for images/files|
|"t"|Text|Default mode 


* open("file.txt","r")   # read
* open("file.txt","w")   # write
* open("file.txt","a")   # append

## 4. Writing Data to a file

write() writes one string to a file.

```python
file = open("C:\\Users\\User\\Desktop\\data.txt",'w')

file.write("Name: Ankit\n")
file.write("Age:22\n")
file.write("Course:Python\n")

file.close()
```

### 5. Writing Multiple Lines(Writelines)

writelines() writes multiple lines at once.


```python
file = open("C:\\Users\\User\\Desktop\\data.txt","w")


lines = [
    "Python is a Programming languages.\n",
    "File Handling\n",
    "Object Oriented Programming\n"
]

file.writelines(lines)
file.close()
```

```python
x = open("C:\\Users\\User\\Desktop\\data.txt",'r')

y = x.read()
print(y)
```

### 6. Using "with" statement(Best Practice)

Instead of manually closing files, Python provices with statement.

Benefits:
* Automatically closes the file.
* safer
* Cleaner code

```python
fileName = "C:\\Users\\User\\Desktop\\data.txt"

with open(fileName,'w') as file:
    file.write("Welcome to Python file Handling")
```

```python
with open(fileName,'r') as f:
    print(f.read())
```

```python
# Ask the user to enter a file name (without extension)
fileName = input("Enter your filename: ")

# Create the full path of the file by joining folder path + filename + .txt extension
fullFileName = "C:\\Users\\User\\Desktop\\" + fileName + ".txt"

# Open the file in write ('w') mode.
# If the file already exists, it will be overwritten.
# 'as f' creates a file object named f
with open(fullFileName, 'w') as f:

    # Ask user how many lines they want to write in the file
    linesNum = int(input("Enter Number of lines to write: "))

    # Variable to count total characters
    count = 0

    # Variable to count vowels
    vow = 0

    # Variable to count consonants
    cons = 0

    # List to store all vowel characters found in the text
    vowelsChar = []

    # List to store all consonant characters found in the text
    consonentChar = []

    # Loop from 1 to number of lines entered by user
    for con in range(1, linesNum + 1):

        # Take content for each line from the user
        content = input("Enter your content here:")

        # String containing all vowels (both lowercase and uppercase)
        vowels = "aeiouAEIOU"

        # Loop through every character in the entered content
        for j in content:

            # Increase total character count
            count = count + 1

            # Check if the character is a vowel
            if j in vowels:
                # Increase vowel count
                vow += 1

                # Store the vowel character in vowel list
                vowelsChar.append(j)

            else:
                # If not vowel, count it as consonant
                cons += 1

                # Store the consonant character in consonant list
                consonentChar.append(j)

        # Create line content with line number
        # Example: 1.Hello World
        content1 = f"{con}.{content}"

        # Write the line into the file
        f.write(content1)

        # Move to next line in file
        f.write("\n")

    # Create summary string for total characters
    countChar = f"Total Number of Characters:{count}"

    # Create summary string for vowels and show vowel list
    vowChar = f"Total Number of Vowels:{vow},{vowelsChar}"

    # Create summary string for consonants and show consonant list
    consChar = f"Total Number of Consonent:{cons},{consonentChar}"

    # Write heading for file details
    f.write("--------------Details of File------------------\n")

    # Write total characters information
    f.write(countChar)

    # Move to next line
    f.write("\n")

    # Write vowel details
    f.write(vowChar)

    # Move to next line
    f.write("\n")

    # Write consonant details
    f.write(consChar)
```
