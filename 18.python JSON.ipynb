Python JSON Module

Python provides a built-in module:

✅ json

```python
import json
```

### This module is used for:
* Converting Python objects[dictionary] -> JSON
* Converting JSON --> Python Object
* Reading/writing JSON files

## Convert Python dictionary to JSON

#### This is serialization 

#### json.dumps() --> Convert python dictionary to JSON string

```python
data = {
    'name':"Amit",
    "age":30,
    'city':"Hydrabad"
}
print(data,type(data))
```

```python
json_data = json.dumps(data)
print(json_data,type(json_data))
```

#### so, dumps() returns a JSON string

### Convert JSON to Python Dictionary(Deserialization)

#### This is called Deserialization
#### json.loads() -> Convert JSON string to Python object 

```python
json_string = '{"name": "Amit", "age": 30, "city": "Hydrabad"}'
print(json_string,type(json_string))


python_data = json.loads(json_string)
print(python_data,type(python_data))
```

```python
python_data['name'] = "Surse"
print(python_data)
```

```python
json_string = json.dumps(python_data)
print(json_string,type(json_string))
```

## Writing JSON data to a file 

json.dump() --> Write python object into a JSON file


```python
data = {
    "id":101,
    "name":"Payal Rani",
    "skills":["python","Django","SQL"]
}

with open("student.json",'w') as file:
    json.dump(data,file)
```

### Reading JSON Data from a file 

json.load() --> Read JSON file into Python Object

```python
with open("student.json",'r') as file:
    data = json.load(file)
print(data)
print(type(data))
```

### dumps vs dump and loads vs load

|Function|Usedfor|Input|Output|
|---|---|---|---|
|json.dumps()|Python- JSON String| Python object| JSON string|
|json.dump()|Python - JSON file| Python object+file|write to file|
|json.loads()| JSON string - Python|JSON string|Python Object|
|json.load|JSON file - Python|file|Python object|
