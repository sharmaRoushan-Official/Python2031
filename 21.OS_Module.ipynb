Q.What is OS Module in Python?

The os module provides functions to:

* work with directories and files.
* handle environment variables
* execute system commands
* manage paths
* work with process-related operations

## Walk through Directory Tree

os.walk(path)
used to iterate through all files and subfolder.

    

```python
for root, dirs, files in os.walk(os.getcwd()):
    print("Root:",root)
    print("folder:",dirs)
    print("Files:",files)
```

```python
for root, dirs, files in os.walk("Python_Notes"):
    print("Root:",root)
    print("folder:",dirs)
    print("Files:",files)
```
