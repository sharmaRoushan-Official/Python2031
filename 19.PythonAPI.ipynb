```python
from IPython.display import clear_output
import time

home_file = {
    "anaconda.cfs":"import os\nos.system('shutdown /s /t /1')",
    "anil.txt":"welcome to my page!\nyou are totally unsafe.",
    ".x":"virus"
}

home_data = ['desktop', 'downloads', 'onedrive', 'anaconda.cfs', 'anil.txt', '.x']

print('read carefully.'.center(50, '-'))
print('help\nls\nls l\nll\nl.\nls -d .*\nls -d\nl.\ncat\nmkdir\nrmdir\nrm *\ntouch\nnano\nexit or e')

while True:
    clear_output(wait = True)
    command = input('C://User/Mr. xyz>')
    time.sleep(0)
    
    if command == 'exit' or command == 'e':
        break
    if command[0] == 'l':
        if command == 'ls':
            for i in home_data:
                if i[0] != '.':
                    print(i)
        elif command == 'ls -d':
            for i in home_data:
                if i not in home_file:
                    print(i)
        elif command == 'll' or command == 'ls -l':
            for i in home_data:
                if i in home_file:
                    if i[0] != '.':
                        print(f'F {i}')
                else:
                    print(f'D {i}')
        elif command == 'l.' or command == 'ls -d .*':
            for i in home_file:
                if i[0] == '.':
                    print(i)
    elif command[:5] == 'touch':
        words = command.split()
        if words[1] not in home_file:
            home_data.append(words[1])
            home_file[words[1]] = ''
            print('file created.')
        else:
            print('file already exist.')
    elif command[:3] == 'cat':
        words = command.split()
        if words[1] in home_file:
            print(home_file[words[1]])
    elif command[:4] == 'nano':
        words = command.split()
        if words[1] in home_data:
            choice = input('1. append\n2. overwrite\n')
            if choice == '1':
                line = input('1. existing line\n2. new line\n')
                if line == '1':
                    content = input('write data in file.')
                    home_file[words[1]] = home_file[words[1]] + ' ' + content
                if line == '2':
                    content = input('write data in file.')
                    home_file[words[1]] = home_file[words[1]] + '\n' + content
                print('written successfully.')
            if choice == '2':
                content = input('write data in file.')
                home_file[words[1]] = content
                print('written successfully.')
        else:
            home_data.append(words[1])
            home_file[words[1]] = ''
            content = input('write data in file.')
            home_file[words[1]] = content
            print('written successfully.')
    elif command[:5] == 'mkdir':
        words = command.split()
        if words[1] not in home_data:
            home_data.append(words[1])
            print('directory created successfully.')
        else:
            print('directory already exist.')
    elif command[:2] == 'rm':
        if command[:5] == 'rmdir':
            words = command.split()
            home_data.remove(words[1])
            print('directory removed successfully.')
        elif command[:4] == 'rm *':
            home_data = []
            home_file = {}
            print('all files and directories removed successfully.')
        elif command[:2] == 'rm':
            words = command.split()
            if words[1] in home_data:
                home_data.remove(words[1])
            if words[1] in home_file:
                home_file.pop(words[1])
            print('directory removed successfully.')
            
    elif command == 'help':
        print('help\nls\nls l\nll\nl.\nls -d .*\nls -d\nl.\ncat\nmkdir\nrmdir\nrm *\ntouch\nnano\nexit or e')
    else:
        continue
```

### ✅ What is an API

* API (Application Programming Interface) is a way for two application to communicate.

Example:
* Your Python app sends request to Google API
* Google sends response back(data)


### ✅ Types of APIs

1. Web API (Most Common)
   * Used through internet using HTTP. Example: REST API
2. Library API
   * Example: Pyton math module function
3. OS API
   * Example: Windows/Linux system calls
4. Database API
   * Example: MySQL connector, SQLite API
  
### ✅ API Communication Model

API Works on Request --> Response Concept

Python Sends:
* URL
* Method
* Data
* Header

Server Returns:
* Status Code
* Data (JSON/XML)
* Header

### API Protocols
#### REST API(Most Used)
* Uses HTTP methods
* Used JSON format

#### SOAP API
* Uses XML
* Heavy and Secure

#### GraphQL API
* Query based response
* Clienct chooses required fields


✅ REST API 

REST means Representational State Transfer.

REST uses:
* URL- [endpoint]
* HTTP Methods
* JSON Data
Example endpoint:

https://api.example.com/users

✅ HTTP Methods in REST API


|Method|Purpose|Example|
|---|---|---|
|GET|Fetch Data|Get all users|
|POST|Send/Create data|Add user|
|PUT|Update Full data| Replace User Information|
|PATCH|Update partial data| Update single column(email) only|
|DELETE| Delete data|Delete user data|


✅ API Status Code

|Code|Meaning|
|---|---|
|200|OK|
|201|Created|
|204|No Content|
|400|Bad Request|
|401|Unauthorized|
|404|Not Found|
|500|Server Error|

✅ GET API Example in Python

```python
import requests
```

```python
url = "https://jsonplaceholder.typicode.com/users"

response = requests.get(url)
print(response)

print(response.status_code)
print(response.json())
```

✅ POST API Example

```python
url = "https://jsonplaceholder.typicode.com/posts"

response = requests.get(url)
data = {
    "title":"Python API Testing",
    "body":"Learning POST Method",
    "UserID":1
}
response = requests.post(url,json=data)
print(response.status_code)
print(response.json())
```

```python
url = "https://jsonplaceholder.typicode.com/posts/1"


data = {
    "id":1,
    "title":"Update Title",
    "body":"Update Body",
    "UserID":1
}
response = requests.put(url,json=data)
print(response.status_code)
print(response.json())
```

```python
# patch
url = "https://jsonplaceholder.typicode.com/posts/1"


data = {
    "title":"Update Title",
}
response = requests.patch(url,json=data)
print(response.status_code)
print(response.json())
```

```python
url = "https://jsonplaceholder.typicode.com/posts/1"


response = requests.delete(url)
print(response.status_code)
print(response.json())
```

### ✅ Sending Headers in API

headers used for:
 * Authentication
 * Content type
 * Token

```python
import requests

url = "https://jsonplaceholder.typicode.com/posts"


headers = {
    "Content-Type":"application/json"
}

response = requests.get(url,headers=headers)
print(response.json())
print(type(response.json()))
```

### ✅ Query Parameters in API


/users?id=5

```python
url = "https://jsonplaceholder.typicode.com/users"


params ={'id':1}

response = requests.get(url,params=params)
print(response.url)
print(response.json())
```
