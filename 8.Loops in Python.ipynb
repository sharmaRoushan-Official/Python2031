## 2Ô∏è‚É£ Baisc syntax of `for` loop


## Explanation:

|`Part`|Meaning|
|--|--|
|`for`| keyword that starts the loop|
|`variable`|stores one item at a time.|
|`in`|Connects varible to sequence|
|`sequence`|List, string, range, tuple, string, dictionary, sets|
|`:`|Indicates start of the loop body- Mark to end the code line|
|Indentation| Code inside loop|
