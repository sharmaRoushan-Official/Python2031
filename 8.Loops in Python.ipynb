## Explanation:

|`Part`|Meaning|
|--|--|
|`for`| keyword that starts the loop|

|`sequence`|List, string, range, tuple, string, dictionary, sets|
|`:`|Indicates start of the loop body- Mark to end the code line|
|Indentation| Code inside loop|

```python
for i in range(1,10,2):
    print(i)
```

```python
for i in range(0,10):
    if i%2==0:
        print(i)
```

```python
# 1,10 using while 

i=1
j=11

while i<j:
    print(i)
    i=i+1
```

```python
# 19

num = 19
i=1
# 19x1=19
# 19x2=38
# 19x3=57

while i<11:
    print(num,"X",i,"=",num*i)
    i+=1
```

```python
num = 19
i=1
while i<11:
    # print(num,"X",i,"=",num*i)
    print("%dX%d=%d"%(num,i,num*i))
    i+=1
```

```python
num = 19
i=1
while i<11:
    # print(num,"X",i,"=",num*i)
    # print("%dX%d=%d"%(num,i,num*i))
    print(f"{num}X{i}={num*i}")
    i+=1
```

```python
# 1-40

i=1
j=41

while i<j:
    # print(i)
    print(f"\t Table of {i}")
    a=1
    b=11
    while a<b:
        # print(i,a,i*a)
        print(f"{i}X{a}={i*a}")
        a+=1
    i+=1
```
