## Naming Conventions in Programming

### Camel Case (`camelCase`)
In this convention, the first word is in lowercase, and the first letter of each subsequent word is capitalized, with no spaces or underscores.

- **Usage:** Commonly used for variable, function, and method names in languages like **JavaScript**, **Java**, and C#.
- **Example:** `numberOfDonuts`, `favePhrase`

---

### Pascal Case (`PascalCase`)
Also known as Upper Camel Case, this style is similar to camel case, but the first letter of every word is capitalized.

- **Usage:** Typically used for naming classes, interfaces, and namespaces in languages such as **C#**, Java, and TypeScript.
- **Example:** `NumberOfDonuts`, `ClassName`

---

### Snake Case (`snake_case`)
Words are separated by underscores (`_`), and all letters are typically in lowercase.

- **Usage:** The standard convention for variable and function names in scripting languages like Python and Ruby.
- **Example:** `number_of_donuts`, `my_variable_name`

```python
import math
class triangle:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def findTriangle(self):
        if (self.a + self.b) > self.c and (self.a + self.c) > self.b and (self.b + self.c) > self.a:
            if self.a == self.b == self.c:
                print("Equilateral Triangle")
                
            elif self.a == self.b or self.a == self.c or self.b == self.c:
                print("Isosceles Triangle")
            else:
                print("Scalene Triangle")
        else:
            print("Not Valid Triangle")

    def findArea(self):
        if (self.a + self.b) > self.c and (self.a + self.c) > self.b and (self.b + self.c) > self.a:
            s = (self.a + self.b + self.c) / 2
            area = math.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))
            print("Area of triangle =", area)
        else:
            print("Cannot find area, not a valid triangle")

tri1 = triangle(6, 6, 6)
tri1.findTriangle()
tri1.findArea()
```

### 1. Inheritance
 * `Inheritance` allows one class to accquire properties and methods of another class.

 * Parent Class/ Base Class

 * Child Class/Derived Class

### `Benefits`
 * __code reuse__
 * __Easy maintenance__
 * __Reduces redundancy__


### A. Single/Simple Inheritance

 * `One Child class inherits from one parent class.`

```python
class Animal:  # parent   # Base class
    def eat(self):
        print("Animal eats food.")


class Dog(Animal):  # child   # Derived Class
    def bark(self):
        print("Dog barks")


dog1 = Dog()
dog1.bark()
dog1.eat()
```

### B. Multiple Inheritance
 A Child class inherits from `multiple parent classes.`

```python
class father:  # parent class / Base class
    def skills(self):  # method of father
        print('Gardening')

        
class mother: # parent class / Base class
    def talent(self):  # method of parent
        print("Cooking")

        
class child(father,mother):  # child class
    def selfProperty(self):  # method of child 
        print("Cricket")

c = child()
c.skills()
c.talent()
c.selfProperty()
```

### C. Multilevel Inheritance
 * A class inherits from a class which already inherited from another class.

```python
class Grandfather:  # Prarent Class/ Base Class
    def land(self):
        print("Owns land")

class Father(Grandfather):  # Child of Grandfather/ Derived Class / Parent of child class
    def house(self):
        print("Own house")

class child(Father):  # Child of Father
    def car(self):
        print('Owns Car')


c = child()
c.land()
c.house()
c.car()
```

### D. Hierarchical Inheritance

 * Multiple child class inherit from `one parent class.`



```python
class Animal:  # Base Class/ Parent
    def eat(self):
        print('Animal eats')

class Dog(Animal):  # Derived class / child of Animal
    def bark(self):
        print("Dog barks")

class Elephant(Animal):   # Derived class / child of Animal
    def trumpet(self):
        print("Elempant Trumpet")

d = Dog()
d.eat()
d.bark()
```

```python
e = Elephant()
e.eat()
e.trumpet()
```

### E. Hybrid Inheritance
* Combination of `two or more types of inheritance.`
* Usually Combines:
* `Multiple`
* `Multilevel`
* `Hierarchical`

```python
class A:  # parent class/ Base Class
    def methodA(self):
        print('Class A')

class B(A):
    def methodB(self):
        print("Class B")
        
class C(A):
    def methodC(self):
        print("Class C")

class D(B,C):
    def methodD(self):
        print("Class D")

d = D()
d.methodA()
d.methodB()
d.methodC()
d.methodD()
```
